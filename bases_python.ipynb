GENERADORES EN PYTHON !!

ESTRUCTURAS QUE EXTRAEN VALORES DE UNA FUNCION Y SE ALMACENAN EN OBJETOS.
ESTOS VALORES SE ALMACENAN DE UNO EN UNO 
ALMACENAN UN VALOR 

VENTAJAS!! 

SON MAS EFICIENTES QUE LAS FUNCIONES TRADICIONALES 
MUY UTILES CON LISTA DE VALORES INFINITOS 
BAJO DETERMINADOS ESCENARIOS, SERA MUY UTIL QUE UN GENERADOR DEVUELVA LOS 
VALORES DE UNO EN UNO.

```python

def generar_pares(limite):
    num = 1 
    lista = []
    while num < limite:
        lista.append(num * 2)
        num += 1 
    return lista

print(generar_pares(10))

```

```python
#UTILIZANDO LA FUNCION YIELD 

def generar_pares(limite):
    num = 1 
    while num < limite:
        yield num * 2 
        num += 1 

devuelve_pares = generar_pares(10)

print(next(devuelve_pares))
print("Siguiente par....")
print(next(devuelve_pares))
print("Siguiente par....")
print(next(devuelve_pares))
print("Siguiente par....")

```

```python
#POO V 
#Encapsulacion de metodos 
#CONTINUACION 3....
import time
class Coche():
 def __init__(self):
  self.largochasis = 250
  self.anchochasis = 120
  self.__ruedad = 4 
  self.enmarcha = False 
 
 def arrancar(self, arrancar):
  self.enmarcha = arrancar
  if self.enmarcha == True:
   chequeo = self.__chequeo_interno()
  if self.enmarcha == True and chequeo == True:
   return "El coche esta en marcha"
  elif self.enmarcha and chequeo == False:
   return "Algo ha ido mal en el chequeo"
  else:
   return "El coche esta parado"

 def estado_del_coche(self):
  print(f"El coche tiene {self.__ruedad}, Ruedas, un ancho de {self.anchochasis} y un largo de {self.largochasis}")

 def __chequeo_interno(self): #ENCAPSULAMOS  PARA QUE SOLO SE PUEDA USAR CUANDO LO LLAMEMOS EN UN ALGORITMO VALIDO Y NO SOLO POR LLAMAR. 
  print("Chequando al Coche....")
  time.sleep(4)
  self.gasolina = "Ok"
  self.aceite = "Ok"
  self.puertas = "Cerradas"

  if (self.gasolina == "Ok" and self.aceite == "Ok" and self.puertas == "Cerradas"):
   return True 
  else:
   return False 


micoche2 = Coche()
micoche = Coche()


print(micoche.arrancar(True))
micoche.estado_del_coche()

print("A CONTINAUCION CREAREMOS EL SEGUNDO OBJETO... ")

print(micoche2.arrancar(False)) 
micoche2.estado_del_coche()

```
