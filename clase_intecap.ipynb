```python
#ORDENAMIENTO DE LISTAS 

# Lista de números desordenados
numeros = [5, 2, 8, 1, 3]

# Ordenar la lista de manera ascendente
numeros.sort()

# Imprimir la lista ordenada
print("Lista original (ordenada):", numeros)
```

```python
#ORDENAMIENTO INVERSA 
numeros = [5, 2, 8, 1, 3]

# Ordenar la lista de manera descendente
numeros_ordenados_descendente = sorted(numeros, reverse=True)

print(numeros_ordenados_descendente)

```

```python
#CODIGO QUE LE USUARIO PIDA NUMEROS ENTEROS PARA GUARDARLO EN UNA LISTA

lista_numeros = []
print(f"La lista actual es: {lista_numeros}")

for agregar in range(1, 10 + 1):
    numero_agregar = input("Ingrese un numero para agregar: ")
    
    if not numero_agregar.isdigit():
        print(f"El numero {numero_agregar} no es positivo!!")
    else:
        numero_agregar = int(numero_agregar)  # por si el usuario se le ocurre agregar numeros con decimales.
        lista_numeros.append(numero_agregar)
        print("Numero Agregado exitosamente!!")
    
print(f"La lista contiene estos números: {lista_numeros}")

 
```

```python
#METODO BURBUJA 
my_list = [8, 10, 6, 2, 41,1,1, 0,1]  # lista a ordenar
swapped = True  # Lo necesitamos verdadero (True) para ingresar al bucle while.

while swapped:
    swapped = False  # no hay intercambios hasta ahora
    for i in range(len(my_list) - 1): #EL -1 ES PARA QUE NO SE SALGA DEL RANGO
        if my_list[i] > my_list[i + 1]:# my_list[i + 1] ES LA SEGUNDA POSICION CON LA QUE SE ESTA COMPARANDO 
            swapped = True  # ¡ocurrió el intercambio!
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]

print(my_list)

```

```python
#PROGRAMA EN ENCONTRAR EL NUMERO MAS GRANDE 

my_list = [17, 3, 11, 5, 1, 9, 7, 15, 13]
largest = my_list[0]

for i in range(1, len(my_list)):
    if my_list[i] > largest:
        largest = my_list[i]

print(largest)
```
