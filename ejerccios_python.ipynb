```python
lista_vocal = ["a", "e", "i", "o", "u"] #HACEMOS UNA LSITA PARA COMPARAR LUEGO CON CONSONANTES Y VOCALES 

def vocal(parametro): #EL PARAMETRO ES EL TEXTO QUE NOS DARAN 
    letra_vocal = False #INICIAMOS CON FALSE PORQUE NO SABEMOS SI ES VOCAL O NO 
    parametro = parametro.replace(" ", "").lower() #REMPLAZAMOS ESPACIOS VACIOS POR ESPACIOS CERRADOS Y CONVERTIMOS EN MINUSCULA #PARA QUE NO HAYA PROBLEMAS EN LA COMPARACION CON LA LISTA 
    if parametro.isalpha(): #ESTA FUNCION ISALPHA VA VERIFICAR QUE SOLO TEXTO VENGA, PARA QUE NO PONGAN NUMEROS
        for es_vocal in parametro:
            if es_vocal in lista_vocal: #SI LA PRIMERA LETRA ES VOCAL ENTRA AL CONDICIONAL Y SE VUELVE TRUE 
                letra_vocal = True 
                print(f"La letra {es_vocal} es {letra_vocal}")
            else: 
                print(f"La letra {es_vocal} No es vocal :D {letra_vocal}") #SI NO ES VOCAL ENTRA AL ELSE Y SIGUE CON EL FALSE
    else:
        print("El parametro contiene numeros :D !! Eh tramposillo") #MENSAJE DE ERRRO POR SI COLOCAN NUMERO CON  EL TEXTO 

while True:
    try: 
        texto = input("Ingrese un texto para ver si hay vocales: ")
        vocal(texto) #LLAMAMOS A LA FUNCION PARA QUE LEEA ESTE TEXTO 
    except Exception as e:
        print(f"Ups encontramos un error {e}") #COMO ULTIMO HACEMOS UNA EXCEPTION PARA VER SI OCURRE ALGUN ERROR MIENTRAS SE EJECUTE 
```

```python
#Escribir una función sum() y una función multip() que sumen y multipliquen respectivamente todos los números de una lista. Por ejemplo: sum([1,2,3,4]) debería devolver 10 y multip([1,2,3,4]) debería devolver 24.

def suma(lista):
    resultado = sum(lista)
    print(f"Suma de los números: {resultado}")

def multi(lista):
    resultado = 1
    for num in lista:
        resultado *= num #AQUI MULTIPLICA LOS NUMEROS 
    print(f"Multiplicación de los números: {resultado}")

intentos = 0
while intentos < 5:
    print("+--- Ingreso de Números ---+")
    mi_lista = []
    for i in range(int(input("Ingrese la cantidad de números que desea agregar: "))): #SE PUEDE HACER ESTO EN PYTHON CON FOR
        numero = int(input("Ingrese un número: "))
        mi_lista.append(numero)

    suma(mi_lista)
    multi(mi_lista)
    break




```

```python
#Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor futuro de la inversión cada año del periodo indicado.
#Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor actual del capital cada año del periodo indicado.

def calculadora_interes_futuras(capital, tipo_interes, numero_año):
  print("VALOR FUTURO")
  for año in range(numero_año):
    valor_futuro = capital * (1 + tipo_interes)
    print(f"Año {año} : {valor_futuro}")
    capital = valor_futuro

def calculadora_interes_actuales(capital, tipo_interes, numero_año):
  print("VALOR ACTUAL")
  for año in range(0, -numero_año, -1):
    valor_actual = capital / (1 + tipo_interes)
    print(f"Año {año} : {valor_actual}")
    capital = valor_actual
  pass 

usuario_capital = str(1000)
banco_interes = 0.10
año_usuario = str(3)

if usuario_capital.isdigit() and año_usuario.isdigit():
   calculadora_interes_futuras(int(usuario_capital), banco_interes, int(año_usuario))
   calculadora_interes_actuales(int(usuario_capital), banco_interes, int(año_usuario))
 
else:
  print("Letra no!!")
```
