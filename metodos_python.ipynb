```python
#La función find(sub[, desde[, hasta]]) en Python se utiliza para encontrar la posición de la primera aparición de la subcadena sub en la cadena principal. Si se incluye el parámetro desde, la búsqueda comienza desde esa posición y si se especifica el parámetro hasta, la búsqueda termina en esa posición. Aquí tienes un ejemplo:

cadena = "Python es un lenguaje de programación"
posicion = cadena.find("un")

print(f"La posición de 'un' en la cadena es: {posicion}")
```

```python
#Si deseas encontrar todas las ocurrencias de una subcadena en una cadena, puedes usar un bucle y la función find de la siguiente manera:

cadena = "Python es un lenguaje de programación. Python es un lenguaje popular."
subcadena = "un"
posicion = cadena.find(subcadena)

while posicion != -1:
    print(f"Se encontró '{subcadena}' en la posición: {posicion}")
    posicion = cadena.find(subcadena, posicion + 1)
```

```python
#La función isalnum() en muchos lenguajes de programación devuelve True si todos los caracteres en la cadena son alfanuméricos (letras o dígitos) y la cadena no está vacía. Devuelve False si la cadena contiene al menos un carácter que no es alfanumérico o si la cadena está vacía.Aquí tienes un ejemplo en Python:

cadena1 = "Hola123"
cadena2 = "Hola Mundo!"
cadena3 = "123456"
cadena4 = ""

resultado1 = cadena1.isalnum()
resultado2 = cadena2.isalnum()
resultado3 = cadena3.isalnum()
resultado4 = cadena4.isalnum()

print(resultado1)  # True
print(resultado2)  # False (espacio y signo de exclamación no son alfanuméricos)
print(resultado3)  # True
print(resultado4)  # False (cadena vacía)


#La función isalnum() es útil cuando necesitas verificar si una cadena contiene únicamente letras y dígitos, por ejemplo, al validar la entrada del usuario o al procesar datos en un programa. SOLO DARA RESULTADOS TRUE O FALSE
```

```python
#Correcto, la función isalpha() también es una función de cadena que se utiliza comúnmente en muchos lenguajes de programación, incluido Python. Devuelve True si todos los caracteres en la cadena son letras y la cadena no está vacía. Si la cadena contiene al menos un carácter que no es una letra o si la cadena está vacía, entonces devuelve False.

cadena1 = "Hola"
cadena2 = "Hola123"
cadena3 = "123"
cadena4 = ""

resultado1 = cadena1.isalpha()
resultado2 = cadena2.isalpha()
resultado3 = cadena3.isalpha()
resultado4 = cadena4.isalpha()

print(resultado1)  # True
print(resultado2)  # False (contiene números)
print(resultado3)  # False (contiene números)
print(resultado4)  # False (cadena vacía)

#es útil cuando necesitas verificar si una cadena contiene solo letras, por ejemplo, al validar nombres o palabras en la entrada del usuario.
```

```python
#Correcto, la función isdigit() es otra función de cadena común en muchos lenguajes de programación, y en el caso de Python. Devuelve True si todos los caracteres en la cadena son dígitos (números) y la cadena no está vacía. Devolverá False si la cadena contiene al menos un carácter que no es un dígito o si la cadena está vacía.

cadena1 = "123"
cadena2 = "123abc"
cadena3 = "abc"
cadena4 = ""

resultado1 = cadena1.isdigit()
resultado2 = cadena2.isdigit()
resultado3 = cadena3.isdigit()
resultado4 = cadena4.isdigit()

print(resultado1)  # True
print(resultado2)  # False (contiene letras)
print(resultado3)  # False (contiene letras)
print(resultado4)  # False (cadena vacía)

#es útil cuando necesitas verificar si una cadena representa un número, por ejemplo, al validar la entrada del usuario para asegurarte de que esté ingresando un número válido.
```

```python
#Correcto, la función islower() es otra función de cadena que se utiliza para verificar si todas las letras en una cadena son minúsculas y si la cadena tiene al menos una letra en minúscula. Devuelve True si se cumplen ambas condiciones y False en caso contrario.

cadena1 = "hola"
cadena2 = "Hola"
cadena3 = "123"
cadena4 = "hola123"
cadena5 = ""

resultado1 = cadena1.islower()
resultado2 = cadena2.islower()
resultado3 = cadena3.islower()
resultado4 = cadena4.islower()
resultado5 = cadena5.islower()

print(resultado1)  # True (todas las letras son minúsculas)
print(resultado2)  # False (hay mayúsculas presentes)
print(resultado3)  # False (no hay letras en la cadena)
print(resultado4)  # True (todas las letras son minúsculas)
print(resultado5)  # False (cadena vacía)

#es útil cuando necesitas verificar si una cadena está compuesta exclusivamente por letras minúsculas.
```

```python
#Correcto, la función isspace() es otra función de cadena que se utiliza para verificar si una cadena es no vacía y si todos sus caracteres son espacios en blanco. Devuelve True si se cumplen ambas condiciones y False en caso contrario.

cadena1 = "    "
cadena2 = "   Hola   "
cadena3 = "Hola Mundo"
cadena4 = ""
cadena5 = "\t\n\r"

resultado1 = cadena1.isspace()
resultado2 = cadena2.isspace()
resultado3 = cadena3.isspace()
resultado4 = cadena4.isspace()
resultado5 = cadena5.isspace()

print(resultado1)  # True (todos los caracteres son espacios)
print(resultado2)  # False (hay letras presentes)
print(resultado3)  # False (hay letras presentes)
print(resultado4)  # False (cadena vacía)
print(resultado5)  # True (todos los caracteres son espacios en blanco)

#es útil cuando necesitas verificar si una cadena está formada solo por caracteres de espacio en blanco, como espacios, tabulaciones o saltos de línea.
```

```python
#Correcto, la función isupper() es una función de cadena que se utiliza para verificar si todas las letras en una cadena son mayúsculas y si la cadena tiene al menos una letra en mayúscula. Devuelve True si ambas condiciones se cumplen y False en caso contrario.

cadena1 = "HOLA"
cadena2 = "Hola"
cadena3 = "123"
cadena4 = "HOLA123"
cadena5 = ""

resultado1 = cadena1.isupper()
resultado2 = cadena2.isupper()
resultado3 = cadena3.isupper()
resultado4 = cadena4.isupper()
resultado5 = cadena5.isupper()

print(resultado1)  # True (todas las letras son mayúsculas)
print(resultado2)  # False (hay minúsculas presentes)
print(resultado3)  # False (no hay letras en la cadena)
print(resultado4)  # True (todas las letras son mayúsculas)
print(resultado5)  # False (cadena vacía)

#es útil cuando necesitas verificar si una cadena está compuesta exclusivamente por letras mayúsculas.
```

```python
#Correcto, la función lstrip() es una función de cadena que se utiliza para eliminar los espacios en blanco (u otros caracteres especificados) del lado izquierdo de la cadena y devuelve una nueva cadena sin esos espacios iniciales.

cadena = "   Hola Mundo   "
nueva_cadena = cadena.lstrip()

print("Original:", repr(cadena))  # repr() se utiliza para mostrar los espacios en blanco (ES OPCIONAL)
print("Nueva:", repr(nueva_cadena))

#Esta función es útil cuando necesitas limpiar o normalizar cadenas que pueden tener espacios en blanco innecesarios al principio. Puedes pasar un argumento opcional a lstrip() para especificar qué caracteres deseas eliminar en lugar de los espacios en blanco por defecto. Por ejemplo, cadena.lstrip('H') eliminará todos los caracteres 'H' al principio de la cadena.
```

```python
#Correcto, la función replace(v, n) en Python devuelve una copia de la cadena original donde todas las apariciones de la subcadena v se han sustituido por la subcadena n. Aquí tienes un ejemplo:

cadena_original = "Hola, mundo. Hola, Python."
nueva_cadena = cadena_original.replace("Hola", "Hola de nuevo")

print("Original:", cadena_original)
print("Nueva:", nueva_cadena)

#Esta función es útil cuando necesitas reemplazar ciertos patrones o subcadenas dentro de una cadena con otra cadena específica.
```

SUBSECUENCIAS EN PYTHON 

Las cadenas son simplemente un tipo especial de secuencias (otro son las listas, que veremos
luego). Todas las secuencias pueden ser “troceadas” utilizando la notaci´on de slices:

```python
c = "DE LA ROSA CHICOJ"
print(c[0]) #BUSCA EN LA CADENA LA POSICION QUE LE INDIQUEMOS Y MUESTRA EL DATOS QUE ESTA AHI 
print(c[-1]) #ESTO ES COMO REVERSA -1 = 1 ES DECIR QUE BUSCAMOS LA PRIMERA POSICION DE LA ULTIMA LETRA DE LA CADENA EJEMPLO EMPIEZA CON J SI USAMOS [-1]
print(c[2:-2]) #Ahi esta lo que hablamos con el [-1] solo que ahorita con dos posiciones a buscar
print(c[3:5]) #un intervalo de posiciones para saber las letras de ese rango simplemente
print(c[:3]) #solo imprime las 3 posiciones en este caso tambien toma el espacio en blanco 
print(c[3:]) #comienza a imprimir desde la posicion 3
#Los n´umeros negativos se consideran posiciones desde el final de la secuencia. Una manera de ver c´omo se elige el subintervalo es la siguiente. Imagina que las posiciones corresponden a marcas entre los caracteres:
```

```python
# Definir algunas variables
cadena = "Hola, soy una cadena"
numero_entero = 42
numero_decimal = 3.14

# Verificar si las variables son instancias de tipos específicos
print(isinstance(cadena, str))  # True si es una cadena (str)
print(isinstance(numero_entero, int))  # True si es un entero (int)
print(isinstance(numero_decimal, float))  # True si es un decimal (float)

# También puedes comprobar múltiples tipos a la vez
print(isinstance(numero_entero, (int, float)))  # True si es int o float

#En este ejemplo, isinstance() se utiliza para verificar si las variables son instancias de los tipos de datos específicos. La función devuelve True si la variable es del tipo indicado y False en caso contrario.
```
